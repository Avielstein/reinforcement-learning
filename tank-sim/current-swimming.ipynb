```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation
import gymnasium as gym
from gymnasium import spaces
import torch
import torch.nn as nn
from stable_baselines3 import PPO
from stable_baselines3.common.env_checker import check_env
from stable_baselines3.common.evaluation import evaluate_policy
from matplotlib.colors import LinearSegmentedColormap
import time

class FishWorldEnv(gym.Env):
    """
    A 2D water environment where a fish (agent) needs to learn to stay in the center
    while dealing with water currents that change over time.
    
    State space: [x_position, y_position, x_velocity, y_velocity, 
                  current_1_x, current_1_y, current_1_strength, ..., 
                  current_n_x, current_n_y, current_n_strength]
    Action space: [thrust_x, thrust_y] - Continuous values between -1 and 1
    """
    
    def __init__(self, tank_size=10.0, max_steps=300, num_currents=3, 
                 current_strength_range=(0.1, 0.8), current_change_prob=0.01):
        super(FishWorldEnv, self).__init__()
        
        # Environment parameters
        self.tank_size = tank_size
        self.max_steps = max_steps
        self.num_currents = num_currents
        self.current_strength_range = current_strength_range
        self.current_change_prob = current_change_prob
        
        # Agent dynamics parameters
        self.mass = 1.0
        self.water_resistance = 0.2
        self.max_force = 1.0
        self.dt = 0.1  # time step
        self.fish_radius = 0.5
        
        # Current dynamics 
        self.current_speed = 0.01
        self.current_radius = 3.0  # Area of effect for currents
        self.currents = []
        
        # Define action and observation spaces
        self.action_space = spaces.Box(
            low=-1.0, high=1.0, shape=(2,), dtype=np.float32
        )
        
        # State includes fish position, velocity, and info about each current
        # [x, y, vx, vy, current1_x, current1_y, current1_strength, ...]
        obs_low = np.array(
            [-self.tank_size, -self.tank_size, -5.0, -5.0] + 
            [-self.tank_size, -self.tank_size, -1.0] * self.num_currents
        )
        obs_high = np.array(
            [self.tank_size, self.tank_size, 5.0, 5.0] + 
            [self.tank_size, self.tank_size, 1.0] * self.num_currents
        )
        self.observation_space = spaces.Box(low=obs_low, high=obs_high, dtype=np.float32)
        
        # Initialize state
        self.state = None
        self.steps = 0
        self.reset()
        
    def _create_currents(self):
        """Create or update random water currents"""
        currents = []
        for _ in range(self.num_currents):
            # Random position
            x = self.np_random.uniform(-0.8 * self.tank_size, 0.8 * self.tank_size)
            y = self.np_random.uniform(-0.8 * self.tank_size, 0.8 * self.tank_size)
            
            # Random strength and direction
            strength = self.np_random.uniform(*self.current_strength_range)
            direction = self.np_random.uniform(0, 2 * np.pi)
            
            currents.append({
                'position': np.array([x, y]),
                'strength': strength,
                'direction': direction,
                'vector': np.array([strength * np.cos(direction), 
                                    strength * np.sin(direction)])
            })
        return currents
        
    def reset(self, seed=None, options=None):
        """Reset the environment to an initial state"""
        super().reset(seed=seed)
        
        # Random initial position
        x = self.np_random.uniform(-0.5 * self.tank_size, 0.5 * self.tank_size)
        y = self.np_random.uniform(-0.5 * self.tank_size, 0.5 * self.tank_size)
        
        # Initialize with zero velocity
        self.fish_state = np.array([x, y, 0.0, 0.0], dtype=np.float32)
        self.steps = 0
        
        # Create water currents
        self.currents = self._create_currents()
        
        # Create full observation
        self.state = self._get_observation()
        
        return self.state, {}
    
    def _get_observation(self):
        """Combine agent state with current information into observation"""
        obs = list(self.fish_state)
        for current in self.currents:
            x, y = current['position']
            strength = current['strength']
            obs.extend([x, y, strength])
        return np.array(obs, dtype=np.float32)
    
    def _calculate_current_effect(self, fish_pos):
        """Calculate the effect of all currents on the fish"""
        total_effect = np.zeros(2)
        
        for current in self.currents:
            # Distance from fish to current center
            distance = np.linalg.norm(fish_pos - current['position'])
            
            # If fish is within current's area of influence
            if distance < self.current_radius:
                # Effect decreases with distance (linearly)
                effect_strength = current['strength'] * (1 - distance / self.current_radius)
                effect_vector = current['vector'] * effect_strength
                total_effect += effect_vector
        
        return total_effect
    
    def _update_currents(self):
        """Randomly update currents over time"""
        for i, current in enumerate(self.currents):
            # Small chance to change direction
            if self.np_random.random() < self.current_change_prob:
                new_direction = self.np_random.uniform(0, 2 * np.pi)
                strength = current['strength']
                self.currents[i]['direction'] = new_direction
                self.currents[i]['vector'] = np.array([
                    strength * np.cos(new_direction),
                    strength * np.sin(new_direction)
                ])
            
            # Move current slightly
            self.currents[i]['position'] += self.current_speed * self.currents[i]['vector']
            
            # Keep currents within bounds
            pos = self.currents[i]['position']
            if not (-0.9 * self.tank_size <= pos[0] <= 0.9 * self.tank_size and
                    -0.9 * self.tank_size <= pos[1] <= 0.9 * self.tank_size):
                # If current is too close to edge, bounce it back
                if abs(pos[0]) > 0.9 * self.tank_size:
                    self.currents[i]['vector'][0] *= -1
                    self.currents[i]['direction'] = np.arctan2(
                        self.currents[i]['vector'][1], 
                        self.currents[i]['vector'][0]
                    )
                if abs(pos[1]) > 0.9 * self.tank_size:
                    self.currents[i]['vector'][1] *= -1
                    self.currents[i]['direction'] = np.arctan2(
                        self.currents[i]['vector'][1], 
                        self.currents[i]['vector'][0]
                    )
                    
                # Adjust position to be within bounds
                self.currents[i]['position'] = np.clip(
                    self.currents[i]['position'],
                    -0.9 * self.tank_size,
                    0.9 * self.tank_size
                )
    
    def step(self, action):
        """Take a step in the environment based on the action"""
        self.steps += 1
        
        # Unpack state
        x, y, vx, vy = self.fish_state
        fish_pos = np.array([x, y])
        
        # Apply thrust action
        thrust_x = np.clip(action[0], -1.0, 1.0) * self.max_force
        thrust_y = np.clip(action[1], -1.0, 1.0) * self.max_force
        thrust = np.array([thrust_x, thrust_y])
        
        # Calculate current effects
        current_effect = self._calculate_current_effect(fish_pos)
        
        # Calculate total force (thrust + currents)
        total_force = thrust + current_effect
        
        # Calculate acceleration (F = ma)
        acceleration = total_force / self.mass - self.water_resistance * np.array([vx, vy]) / self.mass
        
        # Update velocity
        new_velocity = np.array([vx, vy]) + acceleration * self.dt
        vx, vy = new_velocity
        
        # Update position
        x += vx * self.dt
        y += vy * self.dt
        
        # Boundary handling - bounce off walls with energy loss
        if abs(x) > self.tank_size - self.fish_radius:
            x = np.sign(x) * (self.tank_size - self.fish_radius)
            vx = -vx * 0.8  # Lose energy on collision
            
        if abs(y) > self.tank_size - self.fish_radius:
            y = np.sign(y) * (self.tank_size - self.fish_radius)
            vy = -vy * 0.8  # Lose energy on collision
        
        # Update fish state
        self.fish_state = np.array([x, y, vx, vy], dtype=np.float32)
        
        # Update currents
        self._update_currents()
        
        # Calculate reward
        # Main component: negative distance from center (higher when closer to center)
        distance_from_center = np.sqrt(x**2 + y**2)
        
        # Exponential reward function - sharper drop-off as you move from center
        position_reward = 1.5 * np.exp(-0.3 * distance_from_center)
        
        # Penalty for high velocities (to encourage stability)
        velocity_penalty = -0.1 * (vx**2 + vy**2)
        
        # Penalty for using too much thrust (to encourage efficiency)
        thrust_penalty = -0.05 * np.sum(np.square(thrust))
        
        # Combine rewards
        reward = position_reward + velocity_penalty + thrust_penalty
        
        # Update full observation state
        self.state = self._get_observation()
        
        # Check if episode is done
        done = self.steps >= self.max_steps
        
        # Additional info
        info = {
            "distance_from_center": distance_from_center,
            "position": (x, y),
            "velocity": (vx, vy),
            "current_effect": current_effect
        }
        
        return self.state, reward, done, False, info
    
    def render(self):
        """Simple text-based rendering"""
        x, y, vx, vy = self.fish_state
        distance = np.sqrt(x**2 + y**2)
        print(f"Position: ({x:.2f}, {y:.2f}), Distance from center: {distance:.2f}")


def visualize_agent(env, model=None, num_episodes=3, render_mode='trajectory'):
    """
    Visualize agent behavior in the FishWorld environment
    
    Parameters:
    - env: The FishWorld environment
    - model: Trained agent model (if None, will use random actions)
    - num_episodes: Number of episodes to visualize
    - render_mode: Either 'trajectory' for static plots or 'animation' for animated plot
    """
    
    for episode in range(num_episodes):
        if render_mode == 'trajectory':
            # Initialize plotting for trajectory
            fig, ax = plt.subplots(figsize=(10, 10))
            ax.set_xlim(-env.tank_size, env.tank_size)
            ax.set_ylim(-env.tank_size, env.tank_size)
            ax.set_xlabel('X position')
            ax.set_ylabel('Y position')
            ax.set_title(f'Fish Trajectory - Episode {episode+1}')
            
            # Draw tank boundary
            tank = plt.Rectangle((-env.tank_size, -env.tank_size), 
                               2*env.tank_size, 2*env.tank_size, 
                               linewidth=2, edgecolor='blue', facecolor='skyblue', alpha=0.3)
            ax.add_patch(tank)
            
            # Draw center target
            center = plt.Circle((0, 0), 1.0, color='green', alpha=0.3)
            ax.add_patch(center)
            
            # Run episode
            obs, _ = env.reset()
            done = False
            trajectory_x = [env.fish_state[0]]
            trajectory_y = [env.fish_state[1]]
            
            while not done:
                if model:
                    action, _ = model.predict(obs, deterministic=True)
                else:
                    action = env.action_space.sample()
                    
                obs, reward, done, _, info = env.step(action)
                trajectory_x.append(env.fish_state[0])
                trajectory_y.append(env.fish_state[1])
                
                # Visualize currents each step
                for i, current in enumerate(env.currents):
                    # Create current visualization
                    c_x, c_y = current['position']
                    strength = current['strength']
                    direction = current['direction']
                    
                    # Draw current center
                    current_color = plt.cm.Blues(0.5 + 0.5 * strength / env.current_strength_range[1])
                    ax.add_patch(plt.Circle((c_x, c_y), 0.3, color=current_color, alpha=0.7))
                    
                    # Draw current direction
                    arrow_dx = np.cos(direction) * strength * 2
                    arrow_dy = np.sin(direction) * strength * 2
                    ax.arrow(c_x, c_y, arrow_dx, arrow_dy, 
                            head_width=0.3, head_length=0.5, fc='blue', ec='blue', alpha=0.7)
                    
                    # Draw current influence area
                    influence = plt.Circle((c_x, c_y), env.current_radius, 
                                          color='blue', alpha=0.1, fill=True)
                    ax.add_patch(influence)
            
            # Plot trajectory
            ax.plot(trajectory_x, trajectory_y, 'r-', alpha=0.7)
            ax.plot(trajectory_x[0], trajectory_y[0], 'bo', label='Start')
            ax.plot(trajectory_x[-1], trajectory_y[-1], 'ro', label='End')
            ax.legend()
            
            plt.grid(True)
            plt.show()
            
        elif render_mode == 'animation':
            # Create animation
            fig, ax = plt.subplots(figsize=(10, 10))
            ax.set_xlim(-env.tank_size, env.tank_size)
            ax.set_ylim(-env.tank_size, env.tank_size)
            ax.set_xlabel('X position')
            ax.set_ylabel('Y position')
            ax.set_title(f'Fish in Water Currents - Episode {episode+1}')
            
            # Draw tank
            tank = plt.Rectangle((-env.tank_size, -env.tank_size), 
                               2*env.tank_size, 2*env.tank_size, 
                               linewidth=2, edgecolor='blue', facecolor='skyblue', alpha=0.2)
            ax.add_patch(tank)
            
            # Draw center target
            center = plt.Circle((0, 0), 1.0, color='green', alpha=0.3)
            ax.add_patch(center)
            
            # Initialize objects that will be animated
            fish = plt.Circle((0, 0), env.fish_radius, color='orange', alpha=0.9)
            ax.add_patch(fish)
            
            current_centers = []
            current_arrows = []
            current_influences = []
            
            for _ in range(env.num_currents):
                center = plt.Circle((0, 0), 0.3, color='blue', alpha=0.7)
                arrow = ax.arrow(0, 0, 0, 0, head_width=0.3, head_length=0.5, 
                                fc='blue', ec='blue', alpha=0.7)
                influence = plt.Circle((0, 0), env.current_radius, color='blue', alpha=0.1)
                
                ax.add_patch(center)
                ax.add_patch(influence)
                
                current_centers.append(center)
                current_arrows.append(arrow)
                current_influences.append(influence)
            
            # Create trajectory line
            trajectory, = ax.plot([], [], 'r-', alpha=0.5)
            trajectory_x = []
            trajectory_y = []
            
            # Reset environment
            obs, _ = env.reset()
            
            # Function to initialize animation
            def init():
                fish.center = (env.fish_state[0], env.fish_state[1])
                
                for i, current in enumerate(env.currents):
                    c_x, c_y = current['position']
                    current_centers[i].center = (c_x, c_y)
                    current_influences[i].center = (c_x, c_y)
                    
                    # Remove old arrow and create new one
                    current_arrows[i].remove()
                    strength = current['strength']
                    direction = current['direction']
                    arrow_dx = np.cos(direction) * strength * 2
                    arrow_dy = np.sin(direction) * strength * 2
                    current_arrows[i] = ax.arrow(c_x, c_y, arrow_dx, arrow_dy, 
                                               head_width=0.3, head_length=0.5, 
                                               fc='blue', ec='blue', alpha=0.7)
                
                trajectory.set_data([], [])
                return [fish, trajectory] + current_centers + current_influences + current_arrows
            
            # Function to update animation
            def animate(i):
                nonlocal obs
                
                if model:
                    action, _ = model.predict(obs, deterministic=True)
                else:
                    action = env.action_space.sample()
                    
                obs, reward, done, _, info = env.step(action)
                
                # Update fish position
                fish.center = (env.fish_state[0], env.fish_state[1])
                
                # Update trajectory
                trajectory_x.append(env.fish_state[0])
                trajectory_y.append(env.fish_state[1])
                trajectory.set_data(trajectory_x, trajectory_y)
                
                # Update currents
                for i, current in enumerate(env.currents):
                    c_x, c_y = current['position']
                    current_centers[i].center = (c_x, c_y)
                    current_influences[i].center = (c_x, c_y)
                    
                    # Remove old arrow and create new one
                    if hasattr(current_arrows[i], 'remove'):
                        current_arrows[i].remove()
                    
                    strength = current['strength']
                    direction = current['direction']
                    arrow_dx = np.cos(direction) * strength * 2
                    arrow_dy = np.sin(direction) * strength * 2
                    current_arrows[i] = ax.arrow(c_x, c_y, arrow_dx, arrow_dy, 
                                               head_width=0.3, head_length=0.5, 
                                               fc='blue', ec='blue', alpha=0.7)
                
                if done:
                    ani.event_source.stop()
                
                return [fish, trajectory] + current_centers + current_influences + current_arrows
            
            # Create animation
            ani = animation.FuncAnimation(fig, animate, frames=env.max_steps,
                                         init_func=init, blit=False, interval=50)
            
            plt.grid(True)
            plt.show()


def train_agent(env, total_timesteps=500000):
    """Train a PPO agent in the FishWorld environment"""
    model = PPO(
        "MlpPolicy",
        env,
        verbose=1,
        learning_rate=0.0003,
        gamma=0.99,  # discount factor
        n_steps=1024,
        batch_size=64,
        n_epochs=10,
        policy_kwargs=dict(
            net_arch=[dict(pi=[128, 128], vf=[128, 128])],
            activation_fn=nn.Tanh
        )
    )
    
    model.learn(total_timesteps=total_timesteps, progress_bar=True)
    return model


# Create and validate environment
env = FishWorldEnv(tank_size=10.0, max_steps=200, num_currents=4, 
                  current_strength_range=(0.2, 0.8), current_change_prob=0.02)
check_env(env)

# Visualize random agent behavior
print("Visualizing untrained (random) agent behavior...")
visualize_agent(env, model=None, num_episodes=1, render_mode='trajectory')

# Train the agent
print("Training agent...")
model = train_agent(env, total_timesteps=200000)

# Evaluate the trained agent
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
print(f"Mean reward: {mean_reward:.2f} +/- {std_reward:.2f}")

# Visualize trained agent behavior
print("Visualizing trained agent behavior...")
visualize_agent(env, model=model, num_episodes=2, render_mode='trajectory')

# Optionally create an animation of the trained agent
print("Creating animation of trained agent...")
visualize_agent(env, model=model, num_episodes=1, render_mode='animation')
```
